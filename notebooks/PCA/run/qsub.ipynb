```python
def load_words(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        return set(line.strip() for line in f if line.strip())

def compare_and_save(file1, file2, output_file):
    words1 = load_words(file1)
    words2 = load_words(file2)

    only_in_file1 = sorted(words1 - words2)
    only_in_file2 = sorted(words2 - words1)
    in_both = sorted(words1 & words2)

    with open(output_file, 'w', encoding='utf-8') as out:
        out.write(f"Words only in {file1}:\n")
        out.write("\n".join(only_in_file1) + "\n\n")

        out.write(f"Words only in {file2}:\n")
        out.write("\n".join(only_in_file2) + "\n\n")

        out.write("Words in both files:\n")
        out.write("\n".join(in_both) + "\n")

    print(f"Comparison saved to {output_file}")

# Usage example:
compare_and_save('/mnt/archgen/users/skourtanioti/GeorgianProject/PCA/PCA_poplists/PCA_WEurasia.poplist', '/mnt/archgen/users/bianco/Projects/Assyria/PCA/PCA_list.txt', 'comparison_output.txt')
```
