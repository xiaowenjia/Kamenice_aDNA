```python
def calculate_percentage_of_group(haplogroups, target_groups):
    total_count = len(haplogroups)
    percentages = []
    for target_group in target_groups:
        target_count = sum(1 for hg in haplogroups if hg.startswith(target_group))
        percentage = (target_count / total_count) * 100
        percentages.append(percentage)
    return pd.DataFrame({'target_groups': target_groups, 'percentages': percentages})
def main():
    # Example list of Y haplogroups
    Ymeta_LBA = Ymeta[Ymeta['Periodiation'] == 'KNC_LBA']
    haplogroups_LBA = Ymeta_LBA['Y_Haplogroup']
    Ymeta_EIA = Ymeta[Ymeta['Periodiation'] == 'KNC_EIA']
    haplogroups_EIA = Ymeta_EIA['Y_Haplogroup']
    Ymeta_IA = Ymeta[Ymeta['Periodiation'] == 'KNC_IA']
    haplogroups_IA = Ymeta_IA['Y_Haplogroup']

    # Calculate percentage of haplogroups for multiple target groups
    target_groups = ['R1']
    percentages_LBA_df = calculate_percentage_of_group(haplogroups_LBA, target_groups)
    percentages_EIA_df = calculate_percentage_of_group(haplogroups_EIA, target_groups)
    percentages_IA_df = calculate_percentage_of_group(haplogroups_IA, target_groups)

    # First, merge the LBA and EIA results
    combined_df = pd.merge(percentages_LBA_df, percentages_EIA_df, on='target_groups', suffixes=('_LBA', '_EIA'))

    # Then merge the result with the IA data
    combined_df = pd.merge(combined_df, percentages_IA_df, on='target_groups')

    # Rename columns for clarity
    combined_df.columns = ['target_groups', 'percentages_LBA', 'percentages_EIA', 'percentages_IA']

    # Transpose the DataFrame
    transposed_df = combined_df.set_index('target_groups').T

    print(transposed_df)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def calculate_counts_and_percentages(haplogroups, target_groups):
    """
    Count occurrences and calculate percentages of target haplogroups within a given Series.
    """
    total = len(haplogroups)
    counts = haplogroups.value_counts()

    # Count only target groups
    counts_dict = {group: counts.get(group, 0) for group in target_groups}
    
    # Create DataFrame
    df = pd.DataFrame({
        'target_groups': target_groups,
        'count': [counts_dict[g] for g in target_groups],
        'percentage': [(counts_dict[g] / total * 100) if total > 0 else 0 for g in target_groups]
    })
    return df

# Subset data
Ymeta_LBA = Ymeta[Ymeta['Periodiation'] == 'KNC_LBA']
haplogroups_LBA = Ymeta_LBA['Y_Haplogroup']

Ymeta_EIA = Ymeta[Ymeta['Periodiation'] == 'KNC_EIA']
haplogroups_EIA = Ymeta_EIA['Y_Haplogroup']

Ymeta_IA = Ymeta[Ymeta['Periodiation'] == 'KNC_IA']
haplogroups_IA = Ymeta_IA['Y_Haplogroup']

# Define target groups
target_groups = [
    'E1b1b1a1', 'G2a2a1a2', 'G2a2a2b3', 'G2a2b2b1', 'G2a2b-unresolved',
    'R1b1a1b1', 'R1b1a1b2', 'R1b1a1b-unresolved', 'R1b1b2a2'
]

# Get counts and percentages for each period
LBA_df = calculate_counts_and_percentages(haplogroups_LBA, target_groups)
EIA_df = calculate_counts_and_percentages(haplogroups_EIA, target_groups)
IA_df = calculate_counts_and_percentages(haplogroups_IA, target_groups)

# Merge counts first
combined_counts = pd.merge(
    LBA_df[['target_groups', 'count']],
    EIA_df[['target_groups', 'count']],
    on='target_groups',
    suffixes=('_LBA', '_EIA')
)

combined_counts = pd.merge(
    combined_counts,
    IA_df[['target_groups', 'count']],
    on='target_groups'
).rename(columns={'count': 'count_IA'})

# Merge percentages
combined_percentages = pd.merge(
    LBA_df[['target_groups', 'percentage']],
    EIA_df[['target_groups', 'percentage']],
    on='target_groups',
    suffixes=('_LBA', '_EIA')
)

combined_percentages = pd.merge(
    combined_percentages,
    IA_df[['target_groups', 'percentage']],
    on='target_groups'
).rename(columns={'percentage': 'percentage_IA'})

# Combine counts and percentages together
final_combined_df = pd.merge(combined_counts, combined_percentages, on='target_groups')

# Clean up column names
final_combined_df.columns = [
    'target_groups',
    'Count_LBA', 'Count_EIA', 'Count_IA',
    'Percent_LBA', 'Percent_EIA', 'Percent_IA'
]

# Transpose for plotting or comparison
transposed_df = final_combined_df.set_index('target_groups').T

# Display results
print("Final Combined Counts and Percentages:\n", final_combined_df)
print("\nTransposed Table:\n", transposed_df)
```
